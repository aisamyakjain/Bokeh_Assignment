```python
# 2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.


from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare the data
x_data = [1, 2, 3, 4, 5]
y_data = [5, 3, 4, 2, 1]

# Initialize the plot
plot = figure(title='My Bokeh Plot', x_axis_label='X-axis', y_axis_label='Y-axis', plot_width=500, plot_height=400)

# Add glyphs
plot.line(x_data, y_data, line_color='blue', line_width=2)
plot.circle(x_data, y_data, fill_color='red', size=8)

# Customize the plot (optional)
plot.grid.visible = True

# Show the plot
output_notebook()  # For displaying the plot in a Jupyter Notebook
show(plot)  # Show the plot
```
